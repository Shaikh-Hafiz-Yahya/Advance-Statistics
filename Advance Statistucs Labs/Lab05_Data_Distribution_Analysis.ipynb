#### 3. Background Theory 
#### 3.1 What is Data Distribution? 
A data distribution shows how values are spread across a dataset. 
#### Example: 
Student Marks 
45, 50, 55, 60, 70, 80, 90 
The distribution tells us: 
#####  Are most values low or high 
#####  Are values clustered together 
#####  Are there outliers 
Understanding distribution helps ML models understand patterns in data. 
#### 3.4 Variance 
Variance measures how spread out the data is. 
##### Formula 
[ 
Variance = \frac{\sum (x - \mu)^2}{n} 
] 
Where 
#####  (x) = each value 
#####  (\mu) = mean 
## If variance is: 
Small → data values are close to mean 
Large → data values are spread out 
Source: An Introduction to Statistical Learning. 
#### 3.5 Normal Distribution (Gaussian Distribution) 
A normal distribution is a symmetric bell-shaped curve. 
Properties: 
#####  Mean = Median = Mode 
#####  Most values are near the center 
#####  Few values are very high or very low 
Many ML algorithms assume normally distributed data
